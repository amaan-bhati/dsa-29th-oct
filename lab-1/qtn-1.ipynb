```python
# 1. Hello World
print("Hello, World!")

# 2. Variables and Input
name = input("Enter your name: ")
print("Hello, " + name + "!")

# 3. Constants and Comments
PI = 3.14159
# This is a comment

# 4. Data Types and Type Conversion
age = 25
height = 5.9
is_student = True

# 5. Arithmetic Operators and Expressions
result = 10 + 5
print("Result:", result)

# 6. String Manipulation
message = "Python is fun!"
print("Length of message:", len(message))
print("Uppercase:", message.upper())

# 7. Logical Operators
x = True
y = False
print("x AND y:", x and y)
print("x OR y:", x or y)
print("NOT x:", not x)

# 8. List and Indexing
fruits = ["apple", "banana", "orange"]
print("First fruit:", fruits[0])
fruits.append("grape")
print("Fruits:", fruits)

# 9. Conditional Statements
number = 15
if number % 2 == 0:
    print("Even number")
else:
    print("Odd number")

# 10. Looping with For and While
print("Counting from 1 to 5 with a for loop:")
for i in range(1, 6):
    print(i)

print("Counting from 1 to 5 with a while loop:")
count = 1
while count <= 5:
    print(count)
    count += 1
```
