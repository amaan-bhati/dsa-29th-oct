```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python

df = pd.read_csv("india-crime.csv")

```

```python
print(df.columns)
```

```python
#a) Display total number of Kidnapping and Abduction happened in Maharashtra state.

maharashtra_kidnapping = df[(df['STATE/UT'] == 'MAHARASHTRA')]['Kidnapping and Abduction'].sum()
print("Total number of Kidnapping and Abduction in Maharashtra:", maharashtra_kidnapping)
```

```python
# b) Display total number of Kidnapping and Abduction happened in Maharashtra state of RAIGAD district.
raigad_kidnapping = df[(df['STATE/UT'] == 'MAHARASHTRA') & (df['DISTRICT'].str.contains('RAIGAD', case=False))]['Kidnapping and Abduction'].sum()
print("Total number of Kidnapping and Abduction in Raigad, Maharashtra:", raigad_kidnapping)
```

```python
# c) Display heatmap of 7 columns 
columns_of_interest = ['Rape', 'Kidnapping and Abduction', 'Dowry Deaths',
                       'Assault on women with intent to outrage her modesty',
                       'Insult to modesty of Women', 'Cruelty by Husband or his Relatives',
                       'Importation of Girls']
heatmap = df[columns_of_interest].corr(method='spearman')  # Use Spearman correlation for ordinal data
fig, ax = plt.subplots(figsize=(10, 10))
ax.set_title('Heatmap of Correlations')
ax.pcolor(heatmap, vmin=-1, vmax=1, cmap='coolwarm')
plt.show()

```

```python
#d) Display the bar plot Dowry Deaths from 2001 to 2012 in India.

dowry_deaths_india = df.groupby('Year')['Dowry Deaths'].sum().loc[2001:2012]
dowry_deaths_india.plot(kind='bar', figsize=(10, 6))
plt.title('Dowry Deaths in India (2001-2012)')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.show()

```

```python
#e) Display the bar plot Dowry Deaths from 2001 to 2012 in state of west Bengal.
dowry_deaths_west_bengal = df[(df['STATE/UT'] == 'WEST BENGAL') & (df['Year'] >= 2001) & (df['Year'] <= 2012)].groupby('Year')['Dowry Deaths'].sum()
dowry_deaths_west_bengal.plot(kind='bar', figsize=(10, 6), color='green')
plt.title('Dowry Deaths in West Bengal (2001-2012)')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.show()
```
